```python
# LinkedIn Learning Python course by Joe Marini
# Example file for complex types

# Sequences: Lists and Tuples
# These are -- surprise -- sequences of values
mylist = [0, 1, "two", 3.2, False]
print(len(mylist))

# to access a member of a sequence type, use []
print(mylist[2])
print(mylist[-1])
mylist[0] = 10
print(mylist)

# add a list to another list
another_list = [6, 7, 8]
mylist = mylist + another_list
print(mylist)

mystr = "This is a string"
print(mystr[2])

# use slices to get parts of a sequence
print(mylist[1:4]) #Start at index 1, end at index 3, increment by 1 (default)
print(mylist[1:4:2]) #Start at index 1, end at index 3, increment by 2
print(mylist[:4:2]) #Start at index 0, end at 3, increment by 2
print(mylist[1::2]) #Start at index 1, end at last index, increment by 2


# you can use slices to reverse a sequence
print(mylist[::-1])

# Tuples are like lists, but they are immutable
mytuple = (0, 1, 2, "three")
print(mytuple[1])

# Sets are also sequences, but they contain unique values
myset = {1, 2, 3, 2, 4, "hey"}
print(myset)

# Set, however, can not be indexed like lists or tuples
# print(myset[0]) # this will cause an error

# Test for membership
print(1 in mylist)
print(3 in mytuple)
print(5 in myset)
```

```python
# LinkedIn Learning Python course by Joe Marini
# Example file for complex types


# Dictionary: a key-value data structure
mydict = {
    "one" : 1,
    "two" : 2,
    3 : "three",
    4.5 : ["four", "point", "five"]
}

print(mydict)

# dictionaries are accessed via keys
print(mydict["one"])
print(mydict[3])

# you can also set dictionary data by creating a new key
mydict["seven"] = 7
print(mydict)

# Trying to access a nonexistent key will produce an error
# print(mydict["blarg"])

# To avoid this, you can use the "in" operator to see if a key exists
print("two" in mydict)
print("blarg" in mydict)

# You can retrieve all of the keys and values from a dictionary
print(mydict.keys())
print(mydict.values())

# You can also iterate over all the items in a dictionary
for key, val in mydict.items():
  print(key, val)
```

```python
# LinkedIn Learning Python course by Joe Marini
# Example file for working with loops


x = 0

# define a while loop
while x < 5:
  print(x)
  x = x + 1

answer = input("Should I stop?")
while answer != "yes":
  print(answer)
  answer = input("Should I stop?")

# define a for loop
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
for d in days:
  print(d)

# use a for loop over a collection


# use the break and continue statements
for d in days:
  if (d == "Thu"):
    continue
  print(d)

for d in days:
  if (d == "Wed"):
    break
  print(d)


# using the enumerate() function to get an index and an item
for i, d in enumerate(days):
  print(i, d)
```

```python
# @title
class Vehicle:
  def __init__(self, bodystyle, fuel_level=100):
    self.bodystyle = bodystyle
    self.fuel_level = fuel_level

  def drive(self, speed):
    self.mode = "drive"
    self.speed = speed

  def honk(self):
    self.mode = "honk"

  def stop(self):
    self.mode = "stop"

class Car(Vehicle):
  def __init__(self, enginetype, fuel_level = 100):
    super().__init__("Car")
    self.wheels = 4
    self.doors = 4
    self.enginetype = enginetype
    self.fuel_level = fuel_level


  def drive(self, speed):
    super().drive(speed)
    print("Driving my", self.enginetype, "Car at", self.speed, "fuel", self.fuel_level)

  def honk(self):
    super().honk()
    print("Honk")

  def stop(self):
    super().stop()
    print("Stopping my", self.enginetype, "with fuel", self.fuel_level)

class Motorcycle(Vehicle):
  def __init__(self, enginetype, hassidecar):
    super().__init__("Motorcycle")

    if (hassidecar):
      self.wheels = 3
    else:
      self.wheels = 2

    self.doors = 0
    self.enginetype = enginetype

  def drive(self, speed):
    super().drive(speed)
    print("Driving my", self.enginetype, "Motorcycle at", self.speed)

car1 = Car("gas", 50)
car2 = Car("electric", 50)
mc1 = Motorcycle("gas", True)

car1.drive(50)
car1.honk()
car1.stop()
```

```python
# @title
# LinkedIn Learning Python course by Joe Marini
# Example file for working with Exceptions


# Errors can happen in programs, and we need a clean way to handle them
# This code will cause an error because you can't divide by zero:
x = 10 / 0

# Exceptions provide a way of catching errors and then handling them in
# a separate section of the code to group them together
try:
  x = 10 / 0
except:
  print("Error")

# You can also catch specific exceptions
try:
  answer = input("What should I divide 10 by? ")
  num = int(answer)
  print(10/num)
except ZeroDivisionError as e:
  print("You can't divide by 0")
except ValueError as e:
  print("You didn't give me a valid number")
finally:
  print("Finally always runs")
```

```python
# Python code​​​​​​‌‌‌‌​​‌‌​​‌​​​‌‌‌​‌‌​‌​‌‌ below
# Use print("messages...") to debug your solution.

show_expected_result = False
show_hints = False

def is_palindrome(teststr):
  # Your code goes here.
  cleaned_str = "".join(char for char in teststr.lower() if char.isalnum())
  reverse_str = cleaned_str[::-1]

  return cleaned_str == reverse_str
```
